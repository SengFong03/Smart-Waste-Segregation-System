Split Dataset into Train-Test Set

```python
import os
import random
import shutil

def split_yolo_dataset(dataset_path, output_path, train_ratio=0.8):
    # Create output directories
    os.makedirs(os.path.join(output_path, 'train', 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_path, 'train', 'labels'), exist_ok=True)
    os.makedirs(os.path.join(output_path, 'test', 'images'), exist_ok=True)
    os.makedirs(os.path.join(output_path, 'test', 'labels'), exist_ok=True)

    # Get all image files
    images_path = os.path.join(dataset_path, 'images')
    image_files = [f for f in os.listdir(images_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

    # Shuffle files
    random.shuffle(image_files)

    # Split based on ratio
    split_index = int(len(image_files) * train_ratio)
    train_files = image_files[:split_index]
    test_files = image_files[split_index:]

    # Copy files to train folder
    for img_file in train_files:
        # Get base name without extension
        base_name = os.path.splitext(img_file)[0]

        # Copy Image
        shutil.copy2(
            os.path.join(dataset_path,'images',img_file),
            os.path.join(output_path,'train','images',img_file)
        )

        # Copy label if exists
        label_file = f"{base_name}.txt"
        label_path = os.path.join(dataset_path, 'labels', label_file)
        if os.path.exists(label_path):
            shutil.copy2(
                label_path,
                os.path.join(output_path, 'train', 'labels', label_file)
            )

    # Copy files to test folder
    for img_file in test_files:
        # Get base name without extension
        base_name = os.path.splitext(img_file)[0]

        # Copy image
        shutil.copy2(
            os.path.join(dataset_path, 'images', img_file),
            os.path.join(output_path, 'test', 'images', img_file)
        )

        # Copy label if exists
        label_file = f"{base_name}.txt"
        label_path = os.path.join(dataset_path, 'labels', label_file)
        if os.path.exists(label_path):
            shutil.copy2(
                label_path,
                os.path.join(output_path, 'test', 'labels', label_file)
            )

    print(f"Dataset split complete:")
    print(f"Total images: {len(image_files)}")
    print(f"Training images: {len(train_files)} ({len(train_files)/len(image_files)*100:.1f}%)")
    print(f"Testing images: {len(test_files)} ({len(test_files)/len(image_files)*100:.1f}%)")
```

```python
from pathlib import Path

dataset_path = Path("/content/split_dataset/train")  # replace with 'path/to/dataset' for your custom data
labels = sorted(dataset_path.rglob("*labels/*.txt"))  # all data in 'labels'
```

```python
import yaml

yaml_file = "/content/split_dataset/data.yaml"  # your data YAML with data directories and names dictionary
with open(yaml_file, encoding="utf8") as y:
    classes = yaml.safe_load(y)["names"]
cls_idx = sorted(classes.keys())
```

```python
import pandas as pd

index = [label.stem for label in labels]  # uses base filename as ID (no extension)
labels_df = pd.DataFrame([], columns=cls_idx, index=index)
```

```python
from collections import Counter

for label in labels:
    lbl_counter = Counter()

    with open(label) as lf:
        lines = lf.readlines()

    for line in lines:
        # classes for YOLO label uses integer at first position of each line
        lbl_counter[int(line.split(" ")[0])] += 1

    labels_df.loc[label.stem] = lbl_counter

labels_df = labels_df.fillna(0.0)  # replace `nan` values with `0.0`
```

## K-Fold Dataset Split

```python
import random

from sklearn.model_selection import KFold

random.seed(0)  # for reproducibility
ksplit = 5
kf = KFold(n_splits=ksplit, shuffle=True, random_state=20)  # setting random_state for repeatable results

kfolds = list(kf.split(labels_df))
```

```python
folds = [f"split_{n}" for n in range(1, ksplit + 1)]
folds_df = pd.DataFrame(index=index, columns=folds)

for i, (train, val) in enumerate(kfolds, start=1):
    folds_df[f"split_{i}"].loc[labels_df.iloc[train].index] = "train"
    folds_df[f"split_{i}"].loc[labels_df.iloc[val].index] = "val"
```

```python
fold_lbl_distrb = pd.DataFrame(index=folds, columns=cls_idx)

for n, (train_indices, val_indices) in enumerate(kfolds, start=1):
    train_totals = labels_df.iloc[train_indices].sum()
    val_totals = labels_df.iloc[val_indices].sum()

    # To avoid division by zero, we add a small value (1E-7) to the denominator
    ratio = val_totals / (train_totals + 1e-7)
    fold_lbl_distrb.loc[f"split_{n}"] = ratio
```

```python
import datetime

supported_extensions = [".jpg", ".jpeg", ".png"]

# Initialize an empty list to store image file paths
images = []

# Loop through supported extensions and gather image files
for ext in supported_extensions:
    images.extend(sorted((dataset_path / "images").rglob(f"*{ext}")))

# Create the necessary directories and dataset YAML files
parent_path = Path("/content/split_dataset")
save_path = Path(parent_path / f"{datetime.date.today().isoformat()}_{ksplit}-Fold_Cross-val")
save_path.mkdir(parents=True, exist_ok=True)
ds_yamls = []

for split in folds_df.columns:
    # Create directories
    split_dir = save_path / split
    split_dir.mkdir(parents=True, exist_ok=True)
    (split_dir / "train" / "images").mkdir(parents=True, exist_ok=True)
    (split_dir / "train" / "labels").mkdir(parents=True, exist_ok=True)
    (split_dir / "val" / "images").mkdir(parents=True, exist_ok=True)
    (split_dir / "val" / "labels").mkdir(parents=True, exist_ok=True)

    # Create dataset YAML files
    dataset_yaml = split_dir / f"{split}_dataset.yaml"
    ds_yamls.append(dataset_yaml)

    with open(dataset_yaml, "w") as ds_y:
        yaml.safe_dump(
            {
                "path": split_dir.as_posix(),
                "train": "train",
                "val": "val",
                "names": classes,
            },
            ds_y,
        )
```

```python
import shutil

from tqdm import tqdm

for image, label in tqdm(zip(images, labels), total=len(images), desc="Copying files"):
    for split, k_split in folds_df.loc[image.stem].items():
        # Destination directory
        img_to_path = save_path / split / k_split / "images"
        lbl_to_path = save_path / split / k_split / "labels"

        # Copy image and label files to new directory (SamefileError if file already exists)
        shutil.copy(image, img_to_path / image.name)
        shutil.copy(label, lbl_to_path / label.name)
```

```python
import yaml
from pathlib import Path

# Define the old and new base paths
old_base_path = "/content/split_dataset/train/2025-04-24_5-Fold_Cross-val"
new_base_path = "/content/split_dataset/2025-04-24_5-Fold_Cross-val"

# Find all YAML files in the cross-validation directory
cv_dir = Path(new_base_path)
yaml_files = list(cv_dir.glob("**/*.yaml"))

# Update each YAML file
for yaml_path in yaml_files:
    # Read the YAML file
    with open(yaml_path, 'r') as f:
        yaml_data = yaml.safe_load(f)

    # Update the paths
    if 'path' in yaml_data and old_base_path in yaml_data['path']:
        yaml_data['path'] = yaml_data['path'].replace(old_base_path, new_base_path)

    # Update train and val paths if they're absolute
    for key in ['train', 'val']:
        if key in yaml_data and old_base_path in str(yaml_data[key]):
            yaml_data[key] = yaml_data[key].replace(old_base_path, new_base_path)

    # Write the updated YAML back to file
    with open(yaml_path, 'w') as f:
        yaml.dump(yaml_data, f, default_flow_style=False)

    print(f"Updated {yaml_path}")

print(f"Updated {len(yaml_files)} YAML files")
```

```python
from ultralytics import YOLO

results = {}
# Define your training configuration
batch = 16
project = "kfold_waste_detect"
epochs = 30
imgsz = 640  # Add image size
patience = 20  # Add early stopping patience

# Use a pre-trained YOLOv8 model (you can choose size based on your needs)
weights_path = "yolov8n.pt"  # nano model (fastest)
# Other options: yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt

for k, dataset_yaml in enumerate(ds_yamls):
    print(f"\n=== Training Fold {k+1}/{len(ds_yamls)} ===")

    # Initialize a fresh model for each fold
    model = YOLO(weights_path)  # 'task="detect"' is not needed for YOLOv8

    # Train model on this fold
    results[k] = model.train(
        data=dataset_yaml,
        epochs=epochs,
        batch=batch,
        imgsz=imgsz,
        patience=patience,
        project=project,
        name=f"fold_{k + 1}"
    )

    # Validate after training
    val_metrics = model.val()
    print(f"Fold {k+1} validation mAP50-95: {val_metrics.box.map:.4f}")
```
